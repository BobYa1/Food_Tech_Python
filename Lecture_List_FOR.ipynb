**יצירת רשימה בעזר לולאה FOR**

```
x = range(1, 20, 2)
a_list = [] # Declare empty list
b_list = []
c_list = []
for n in x:
  print(n)
  a_list.append(n)      # לעשות ככה זה נכון
  b_list = [b_list , n] # לא לעשות ככה
  c_list = n # לא לעשות ככה

print(a_list)
print(b_list)
print(c_list)

# a_list לא שקול b_list לא שקול c_list
```

```
x = range(1, 20, 2)
a_list = [] # Declare empty list

for n in x:
  print(n)
a_list.append(n)      # לעשות ככה זה נכון


print(a_list)

# Correct
```

```
x = range(1, 20, 2)
a_list = [] # Declare empty list

for n in x:
  print(n)
  a_list.append[n]      # לעשות ככה זה נכון


print(a_list)

# Correct
```
